```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df1=pd.read_csv("FB (1).csv")
df1.tail()
```

```python
del df1["High"]
del df1["Low"]
del df1["Open"]
del df1["Adj Close"]
del df1["Volume"]
```

```python
df_twiter=pd.read_csv("daily_score.csv")
df_twiter.head()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df_twiter['date']  = pd.to_datetime(df_twiter['date'])
```

```python
#y = df_twiter[df_twiter['predictions'].resample('D').mean()]
#y.predictions
```

```python
df3 = pd.concat([df1.set_index('Date'), 
                 df_twiter.set_index('date')], 
                 axis=1, 
                 keys=('a', 'b'))
df3.columns = df3.columns.map('_'.join)
df3
```

```python
df3=df3[64:-1]
df3
```

```python
for i in range(1,len(df3)-1):
    if pd.isnull(df3.a_Close[i]):
        if pd.isnull(df3.a_Close[i+1]):
            if pd.isnull(df3.a_Close[i+2]):
                df3.a_Close[i]=(df3.a_Close[i-1]+df3.a_Close[i+3])/2
            else:
                df3.a_Close[i]=(df3.a_Close[i-1]+df3.a_Close[i+2])/2
        else:
            df3.a_Close[i]=(df3.a_Close[i-1]+df3.a_Close[i+1])/2
```

```python
df3.head()
```

```python
#df3["sentiment"].isnull()
nan_rows = df3[df3.isnull().T.any().T]
nan_rows
```

```python
#creat new column of the same size 
change=[0]*len(df3.a_Close)
momentum=[0]*len(df3.a_Close)
decision=[0]*len(df3.a_Close)
#calculate momentum and change
for i in range(1,len(df3.a_Close)):
    if df3.a_Close[i]>df3.a_Close[i-1]:
        momentum[i]='1'
        change[i]=(df3.a_Close[i]-df3.a_Close[i-1])/df3.a_Close[i-1]
    else:
        momentum[i]='0'
        change[i]=(df3.a_Close[i-1]-df3.a_Close[i])/df3.a_Close[i-1]
df3['change']=change
df3['momentum']=momentum
df3.tail()
```

```python
df3.rename(columns={"a_Close":"close","b_pos":"positive","b_neg":"negative","b_class":"sentiment"},inplace=True)
df3.head()
```

```python
df3.momentum=df3.momentum.astype(int)
df3.sentiment=df3.sentiment.astype(int)
```

```python
decision=[0]*len(df3.close)
for i in range(0,len(df3.close)):
    if (df3.momentum[i]==1.0) and (df3.sentiment[i]==1.0):
        decision[i]="buy"
    else:
        decision[i]="sell"

df3['decision']=decision
df3.tail()
```

```python
df3.to_csv("sellorbuy.csv",index=True,header=True)
```

```python
from sklearn.svm import SVC
import numpy as np
from sklearn.cross_validation import train_test_split
```

```python
testdataset =df3.iloc[0:,0:6].values
test_target=df3.iloc[0:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(testdataset, test_target, test_size = 0.30, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting classifier to the Training set
clf = SVC(kernel="rbf", decision_function_shape='ovo')
clf.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = clf.predict(X_test)
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
